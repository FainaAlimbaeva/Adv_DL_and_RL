### Часть первая: крестики-нолики при помощи Q-обучения

В коде, прилагающемся к последней лекции про обучение с подкреплением, реализован Environment для крестиков-ноликов, в котором можно при инициализации указывать разные размеры доски и условия победы, а также функции для рисования, в том числе с указанием оценки различных действий. С этим окружением все задания и связаны.

* Реализуйте обычное (табличное) Q-обучение. Обучите стратегии крестиков и ноликов для доски 3х3.
* Попробуйте обучить стратегии крестиков и ноликов для доски 4х4 и/или 5х5.


```python
class Strategy():
    def __init__(self, epsilon, alpha, gamma, crosses):
        self.Q = {}
        self.epsilon = epsilon
        self.alpha = alpha
        self.gamma = gamma
        self.crosses = crosses
        self.s = None
        self.a = None

    def getActionGreedy(self, s, num_actions):
        if s not in self.Q:
            self.Q[s] = np.zeros(num_actions)
        return np.argmax(self.Q[s]) if np.random.rand() > self.epsilon else np.random.randint(num_actions)
    
    def getAction(self, s, num_actions):
        if s not in self.Q:
            return np.random.randint(num_actions)
        return np.argmax(self.Q[s])
        
    def updateQ(self, s, a, next_s, reward, num_actions):
        gamma_term = self.gamma * np.max(self.Q[next_s]) if next_s in self.Q else 0
        if self.crosses:
            self.Q[s][a] += self.alpha * (reward + gamma_term - self.Q[s][a])
        else:
            self.Q[s][a] += self.alpha * (-reward + gamma_term - self.Q[s][a])  
    
id_turn = {1: 0, -1: 1}

def q_episode(env, agent_X, agent_O):
    done = False
    env.reset()
    prev_s, prev_a = [None, None], [None, None]
    while not done:
        turn = env.curTurn 
        state, actions = env.getHash(), env.getEmptySpaces()
        if turn == 1:
            a_idx = agent_X.getActionGreedy(state, len(actions))
            observation, reward, done, info = env.step(actions[a_idx])
            next_state = observation[0]
            
            if prev_s[id_turn[-turn]] is not None:
                agent_O.updateQ(prev_s[id_turn[-turn]], prev_a[id_turn[-turn]], \
                                next_state, reward, len(actions))
                
            prev_s[id_turn[turn]] = state
            prev_a[id_turn[turn]] = a_idx
            
        else:
            a_idx = agent_O.getActionGreedy(state, len(actions))
            
            observation, reward, done, info = env.step(actions[a_idx])
            next_state = observation[0]
            if prev_s[id_turn[-turn]]  is not None :
                agent_X.updateQ(prev_s[id_turn[-turn]], prev_a[id_turn[-turn]], \
                                next_state, reward, len(actions))
            
            prev_s[id_turn[turn]] = state
            prev_a[id_turn[turn]] = a_idx

    return reward

def Q_learning(env, agent_X, agent_O, N_episodes):
    crosses, naughts = [], []
    for t in tqdm(range(N_episodes)):
        reward = q_episode(env, agent_X, agent_O)
        
        if t % (N_episodes * 0.01) == 0:
            mean_reward_X, _ = evaluate(env, agent_X, None, 200)
            _, mean_reward_O = evaluate(env, None, agent_O, 200)

            crosses.append(mean_reward_X)
            naughts.append(mean_reward_O)
            steps.append(t)
    return crosses, naughts, steps


def play_game(env, agent_X=None, agent_O=None):
    done = False
    env.reset()
    state, actions, turn = env.getState()

    while not done:
        if turn == 1:
            a_idx = get_and_print_move(env, agent_X, state, actions)
        else:
            a_idx = get_and_print_move(env, agent_O, state, actions)

        action = actions[a_idx]
        (state, actions, turn), reward, done, _  = env.step(action)
        
    return reward, turn

        
        
def evaluate(env, agent_X=None, agent_O=None, n_episodes=200):
    history = []
    for _ in range(n_episodes):
        reward, turn = play_game(env, agent_X, agent_O)
        if reward == 1:
            history.append([1, 0])
        elif reward == -1:
            history.append([0, 1])
        else:
            history.append([0, 0])
            
    mean_rewards = sum(np.array(history)) / n_episodes
    return mean_rewards


def plot_history(crosses, naughts, steps, N_ROWS, N_COLS):
    plt.figure(figsize=(12, 6))
    plt.plot(steps, crosses, label='crosses vs random')
    plt.plot(steps, naughts, label='naughts vs random')  
    plt.legend()
    plt.title(f'{N_ROWS} x {N_COLS} Game')
    plt.xlabel('Episodes')
    plt.ylabel('Average wins')
    plt.grid(True)
    plt.show()
```

```python
N_episodes = 100000
alpha = 0.001
epsilon = 0.5
gamma = 1.0
N_ROWS = 3
N_COLS = 3
N_WIN = 3

env = TicTacToe(n_rows=N_ROWS, n_cols=N_COLS, n_win=N_WIN)
agent_X = Strategy(epsilon, alpha, gamma, crosses=True)
agent_O = Strategy(epsilon, alpha, gamma, crosses=False)

crosses, naughts, steps = Q_learning(env, agent_X, agent_O, N_episodes)
```

```python
plot_history(crosses, naughts, steps, N_ROWS, N_COLS)
```

### Часть вторая: добавим нейронных сетей
Реализуйте нейронную сеть для метода DQN на доске для крестиков-ноликов. Не буду ограничивать фантазию, но кажется, что свёртки 3х3 здесь должны неплохо работать (в том числе обобщаться на доски размера побольше).
* Реализуйте DQN с нейронной сетью, обучите стратегии крестиков и ноликов. Замечание: скорее всего, experience replay потребуется сразу же.
* Реализуйте Double DQN и/или Dueling DQN.


```python
class ReplayMemory():
    def __init__(self, capacity):
        self.capacity = capacity
        self.memory = []
        self.position = 0

    def store(self, exptuple):
        if len(self.memory) < self.capacity:
            self.memory.append(None)
        self.memory[self.position] = exptuple
        self.position = (self.position + 1) % self.capacity
       
    def sample(self, batch_size):
        return random.sample(self.memory, batch_size)
    
    def __len__(self):
        return len(self.memory)
```

```python
class QNetwork3x3(nn.Module):
    def __init__(self, layer_size=64):
        nn.Module.__init__(self)
        self.conv = nn.Conv2d(3, layer_size * 2, (3, 3))
        self.fc1 = nn.Linear(layer_size * 2, layer_size)
        self.fc2 = nn.Linear(layer_size, 9)

    def forward(self, x):
        x = F.relu(self.conv(x))
        x = torch.flatten(x, 1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
    
```

```python
id_turn = {1: 0, -1: 1}

device = torch.device("cuda") if torch.cuda.is_available() else torch.device('cpu')

class DQN:
    def __init__(self,  N_ROWS, N_COLS, N_WINS, model=QNetwork3x3,):
        self.n_rows = N_ROWS
        self.n_cols = N_COLS
        self.n_wins = N_WINS
        self.env = TicTacToe(n_rows=self.n_rows, n_cols=self.n_cols, n_win=self.n_wins)
        self.model = [model().to(device), model().to(device)]
        self.memory = [ReplayMemory(100000), ReplayMemory(100000)]
        self.optimizer = [optim.Adam(self.model[0].parameters(), lr=3e-4), 
                          optim.Adam(self.model[1].parameters(), lr=3e-4)]
        self.prev_s = [None, None]
        self.prev_a = [None, None]
        
        self.gamma = 0.8
        self.batch_size = 512
        
        self.eps_init, self.eps_final, self.eps_decay = 0.9, 0.05, 10000
        self.num_step = 0
    
    def to_tensor(self, s):
        size = 3
        crosses = np.array([1 if i == '2' else 0 for i in s]).reshape(size, size)
        noughts = np.array([1 if i == '0' else 0 for i in s]).reshape(size, size)
        spaces = np.array([1 if i == '1' else 0 for i in s]).reshape(size, size)
        return torch.Tensor(np.stack([crosses, noughts, spaces])).reshape(3, size, size)


    def select_greedy_action(self, state, turn):
        return self.model[id_turn[turn]](state.unsqueeze(0)).data.max(1)[1].view(1, 1)

    def select_action(self, state, turn):
        sample = random.random()
        self.num_step += 1
        eps_threshold = self.eps_final + (self.eps_init - self.eps_final) * np.exp(-1. * self.num_step / self.eps_decay)
        if sample > eps_threshold:
            return self.select_greedy_action(state, turn)
        else:
            return torch.tensor([[random.randrange(self.n_rows * self.n_cols)]], dtype=torch.int64)
        
    def run_episode(self, do_learning=True, greedy=False):
        self.env.reset()
        done = False
        self.prev_s = [None, None]
        self.prev_a = [None, None]
        state, actions, turn = self.env.getState()
        while not done:
            state_tensor = self.to_tensor(state)
            with torch.no_grad():
                if greedy:
                    a_idx = self.select_greedy_action(state_tensor.to(device), turn).cpu()
                else:
                    a_idx = self.select_action(state_tensor.to(device), turn).cpu()
            
            self.prev_s[id_turn[turn]] = state_tensor
            self.prev_a[id_turn[turn]] = a_idx
            action = self.env.action_from_int(a_idx.numpy()[0][0])
            observation, reward, done, _ = self.env.step(action)
            next_state, actions, turn = observation
            next_state_tensor = self.to_tensor(next_state)
            if reward == -10:
                transition = (state_tensor, a_idx, next_state_tensor, torch.tensor([reward], dtype=torch.float32))
                self.memory[id_turn[turn]].store(transition)
            else:
                if self.prev_s[id_turn[turn]] is not None:
                    transition = (self.prev_s[id_turn[turn]], self.prev_a[id_turn[turn]], next_state_tensor, torch.tensor([turn*reward], dtype=torch.float32))
                    self.memory[id_turn[turn]].store(transition)
                if self.prev_s[id_turn[-turn]] is not None:
                    transition = (self.prev_s[id_turn[-turn]], self.prev_a[id_turn[-turn]], next_state_tensor, torch.tensor([-turn*reward], dtype=torch.float32))
                    self.memory[id_turn[-turn]].store(transition)
                
            if do_learning:
                self.learn(turn)
            state = next_state

    def learn(self, turn):
        if min(len(self.memory[0]), len(self.memory[1])) < self.batch_size:
            return

        # берём мини-батч из памяти
        transitions = self.memory[id_turn[turn]].sample(self.batch_size)
        batch_state, batch_action, batch_next_state, batch_reward = zip(*transitions)

        batch_state = Variable(torch.stack(batch_state).to(device))
        batch_action = Variable(torch.cat(batch_action).to(device))
        batch_reward = Variable(torch.cat(batch_reward).to(device))
        batch_next_state = Variable(torch.stack(batch_next_state).to(device))

        # считаем значения функции Q
        Q = self.model[id_turn[turn]](batch_state).gather(1, batch_action).reshape([self.batch_size])

        # оцениваем ожидаемые значения после этого действия
        Qmax = self.model[id_turn[turn]](batch_next_state).detach().max(1)[0]
        Qnext = batch_reward + (self.gamma * Qmax)

        # и хотим, чтобы Q было похоже на Qnext -- это и есть суть Q-обучения
        loss = F.smooth_l1_loss(Q, Qnext)

        self.optimizer[id_turn[turn]].zero_grad()
        loss.backward()
        self.optimizer[id_turn[turn]].step()

def test_strategy(dqn, player, n_games):
    total_reward = 0
    for i in range(n_games):
        dqn.env.reset()
        state, actions, turn = dqn.env.getState()
        done = False
        while not done:
            if turn == player:
                a_idx = dqn.select_greedy_action(dqn.to_tensor(state).to(device), player)
                a = dqn.env.action_from_int(a_idx)
            else:
                a_idx = np.random.randint(len(actions))
                a = actions[a_idx]
            observation, reward, done, _ = dqn.env.step(a)
            state, actions, turn = observation
        if reward != -10 and reward == player:
            total_reward += 1
    return total_reward / n_games
    

def DQNLearning(dqn, N_episodes):
  crosses = []
  naughts = []
  steps = []
  for t in tqdm(range(N_episodes)):
      dqn.run_episode()
      if t % (N_episodes * 0.01) == 0:
          cross_res = test_strategy(dqn, player=1, n_games=500)
          naught_res = test_strategy(dqn, player=-1, n_games=500)
          crosses.append(cross_res)
          naughts.append(naught_res)
          steps.append(t)
  return crosses, naughts, steps



```

```python
N_ROWS, N_COLS, N_WINS = 3, 3, 3
N_episodes = 10000

dqn = DQN(N_ROWS, N_COLS, N_WINS)
crosses, naughts, steps = DQNLearning(dqn, N_episodes)
plot_history(crosses, naughts, steps, N_ROWS, N_COLS)
```

```python
class DuelingNetwork(nn.Module):
    def __init__(self, layer_size=128):
        nn.Module.__init__(self)
        self.conv = nn.Conv2d(3, layer_size * 2, (3, 3))
        self.fc1 = nn.Linear(layer_size * 2, layer_size)
        self.fc2 = nn.Linear(layer_size, 1)
        self.fc3 = nn.Linear(layer_size, 9)

    def forward(self, x):
        x = F.relu(self.conv(x))
        x = torch.flatten(x, 1)
        x = F.relu(self.fc1(x))
        value = self.fc2(x)
        adv = self.fc3(x)
        x = value + adv - torch.mean(adv, dim=1, keepdim=True)
        return x
```

```python
N_ROWS, N_COLS, N_WINS = 3, 3, 3
N_episodes = 10000

dueling_dqn = DQN(N_ROWS, N_COLS, N_WINS, model=DuelingNetwork)
crosses, naughts, steps = DQNLearning(dueling_dqn, N_episodes)
plot_history(crosses, naughts, steps, N_ROWS, N_COLS)
```
